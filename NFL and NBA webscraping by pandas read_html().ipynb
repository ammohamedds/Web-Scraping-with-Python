## NFL Football 

```python
import pandas as pd
```

```python
url = 'https://www.pro-football-reference.com/years/2018/'   # i will retrive just one year 
```

```python
df = pd.read_html(url, header = 0)
df
```

### Number of Tables

```python
len(df)  # Three Tables 
```

```python
df_2018= df[0] # show th first table
df_2018
```

### We can see that the table header is repeated so we need to remove it 

###  Data Cleaning

To should create new column and add these header whether AFC EASt, AFC North ,,,,,, then removing this header 

Add New Column for location

```python
# https://www.geeksforgeeks.org/create-a-column-using-for-loop-in-pandas-dataframe/
Location = []
for value in df_2018["W"]:
    if value == "AFC East":
        Location.append("AFC East") 
        value2 = "AFC East"
    elif value == "AFC North":
        Location.append("AFC North")
        value2 = "AFC North"
    elif value == "AFC South":
        Location.append("AFC South")
        value2 = "AFC South"
    elif value == "AFC West":
        Location.append("AFC West")
        value2 = "AFC West"
    else:
        Location.append(value2)
#     print(value)

Location
df_2018["Location"] = Location   
df_2018
```
